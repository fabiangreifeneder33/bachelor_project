```python
from bz_functions import *
from math import pi
from BZ import BZ
```

```python
# set number of interpolation points, interval and a curve
interval = [0, 6 * pi]
n = 19
curve = helix
```

```python
# determine interpolation points equi-spaced in the parameter space
interpolation_points = []
S = np.linspace(interval[0], interval[1], n)
for s in S:
    interpolation_points += [curve(s)]
```

```python
 # set-up Bézier-representation
bz_repr = BZ(interpolation_points)
```

```python
# apply Richardson-method to Bézier curve until there is no more improvement to the max-error
while bz_repr.iter < 200:
    bz_repr.gradient_descent_iter()
```

```python
# plot computed interpolating bezier-curve
bz_repr.plot_results(curve, *interval)
bz_repr.plot_ip_points()
```
